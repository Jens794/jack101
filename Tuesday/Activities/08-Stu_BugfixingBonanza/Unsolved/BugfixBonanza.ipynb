```python
# Create a reference to the CSV and import it into a Pandas DataFrame
import pandas as pd #name error means always add import pandas as pd
csv_path ="Resources/EclipseBugs.csv"
eclipsebugs_df= pd.read_csv(csv_path)# define the data frame meaning give it a name
eclipsebugs_df.head()
```

```python
eclipsebugs_df.columns #its better always check the columns names,so for that call the column function
```

```python
eclipsebugs_df = eclipsebugs_df.rename(columns={'Number of\nComments': 'Number of Comments', 'Bug\nID': 'Bug ID','Assignee\nReal\nName':'Assignee Real Name','Reporter\nReal\nName':'Reporter Real Name','Target\nMilestone':'Target Milestone'})
```

```python
# Finding the average number of comments per bug
average_comments = eclipsebugs_df["Number of Comments"].mean()
average_comments
```

```python
# Grouping the DataFrame by "Assignee"
assignee_group = eclipsebugs_df.groupby("Assignee")

# Count how many of each component Assignees worked on and create DataFrame
assignee_work = pd.DataFrame(assignee_group["Component"].value_counts())
assignee_work.head()
```

```python
# Rename the "Component" column to "Component Bug Count"
eclipsebugs_df = eclipsebugs_df.rename(columns={'Component': 'Component Bug Count'})
```

```python
# Find the percentage of bugs overall fixed by each Assignee
total_bugs = assignee_group["Assignee"].count()
bugs_per_user = assignee_group["Assignee"].value_counts()

user_bug_percent = pd.DataFrame((bugs_per_user/total_bugs)*100)
user_bug_percent.head()
```

```python
# Merge the "Percent of Total Bugs Assigned" into the DataFrame
assignee_work = assignee_work.merge(user_bug_percent, on="Assignee")

# Remove the extra columns
assignee_work = assignee_work[[
    "Assignee", "Percent of Total Bugs", "Component", "Component Bug Count"]]
assignee_work.head()
```
